Câu 4: Python hỗ trợ nhiều kiểu dữ liệu cơ bản, bao gồm:

Kiểu số (Numeric Types):
+ int: Số nguyên (ví dụ: 10, -5, 0)
+ float: Số thực (số có dấu phẩy động) (ví dụ: 3.14, -0.5, 2.0)
+ complex: Số phức (ví dụ: 1 + 2j)

Kiểu chuỗi (String Type):
+ str: Chuỗi ký tự (ví dụ: "hello", 'Python')

Kiểu Boolean:
+ bool: Kiểu logic, chỉ có hai giá trị True hoặc False

Kiểu Sequence Types:
+ list: Danh sách có thứ tự và có thể thay đổi (ví dụ: [1, 2, 3], ['a', 'b'])
+ tuple: Bộ (tuple) có thứ tự và không thể thay đổi (ví dụ: (1, 2, 3), ('a', 'b'))
+ range: Dãy số không thay đổi (thường dùng trong vòng lặp for) (ví dụ: range(5))

Kiểu Set Types:
+ set: Tập hợp không có thứ tự và không chứa phần tử trùng lặp (ví dụ: {1, 2, 3})
+ frozenset: Tập hợp không có thứ tự, không chứa phần tử trùng lặp và không thể thay đổi (ví dụ: frozenset({1, 2, 3}))

Kiểu Mapping Type:
+ dict: Từ điển (dictionary), lưu trữ dữ liệu dưới dạng cặp khóa-giá trị (key-value) (ví dụ: {'a': 1, 'b': 2})

Kiểu Binary Types:
bytes: Dãy byte bất biến (ví dụ: b'hello')
bytearray: Dãy byte có thể thay đổi
memoryview: "View" bộ nhớ của các đối tượng khác

Câu 5: Trong Python, có hai loại ghi chú (comments) chính:

Ghi chú đơn dòng (Single-line comments):
+ Bắt đầu bằng ký tự #.
+ Bất kỳ văn bản nào sau # trên cùng một dòng đều được coi là ghi chú và bị trình thông dịch Python bỏ qua.

Thường được sử dụng để giải thích một dòng mã cụ thể hoặc một phần của mã.

# Đây là một ghi chú đơn dòng
x = 10 # Gán giá trị 10 cho biến x

Ghi chú đa dòng (Multi-line comments) hoặc Docstrings:
+ Mặc dù Python không có cú pháp ghi chú đa dòng riêng biệt như một số ngôn ngữ khác, người ta thường sử dụng chuỗi ký tự đa dòng (triple-quoted strings) làm ghi chú đa dòng hoặc, quan trọng hơn, làm docstrings.
+ Docstrings (chuỗi tài liệu) được sử dụng để tài liệu hóa các module, class, hàm và phương thức. Chúng được đặt ngay sau dòng định nghĩa của đối tượng mà chúng tài liệu hóa.
+ Docstrings có thể được truy cập thông qua thuộc tính __doc__ của đối tượng.

"""
